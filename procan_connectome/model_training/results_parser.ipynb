```python
import pandas as pd 
from sklearn.metrics import accuracy_score
import os 
from procan_connectome.config import DATA_PATH, PLOT_PATH
from procan_connectome.eda.EDA import confusion_matrix_plot
from sklearn.metrics import confusion_matrix
```

```python
results_df = pd.read_csv(os.path.join(DATA_PATH, 'logs', '2021_06_30_21_04_LinearSVC_LOOCV_results.csv'))
cv_df = pd.read_csv(os.path.join(DATA_PATH, 'logs', '2021_07_08_10_50_RF_LOOCV_grid_search.csv'))
```

```python
results_df
```

```python
cv_df['Best_Params'].groupby(cv_df['Best_Params']).count()
```

```python
labels = results_df['y_true'].unique()
cm = confusion_matrix(results_df['y_true'], results_df['y_pred'], labels=labels)
fig = confusion_matrix_plot(cm, labels, fig_title= "Top 0.1% SVC Confusion Matrix", plot_path=PLOT_PATH)
```
